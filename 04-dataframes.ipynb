# Cómo crear un Data Frame

```python
df = pd.DataFrame({
    'A' : np.random.rand(10), #rand genera por defecto números aleatorios entre 0 y 1
    'B' : np.random.rand(10),
    'C' : np.random.rand(10)
}) #Para generar el data frame, usamos la función pd.DataFrame y debemos declararla como diccionario.
```

```python
df
```

```python
df.mean() #Por defecto es la media (sumando los valores en cada columna)
```

```python
df.mean(axis='columns') #La media (sumando los valores de cada fila)
```

```python
df2 = pd.DataFrame({
    'key' : ['A', 'B', 'C', 'A', 'B', 'C'],
    'values' : range(6)
},
columns = ['key', 'values']
)
df2
```

```python
df2.groupby('key')
```

```python
df2.groupby('key').mean() #Lo agrupo por clave, 
#entonces Python tomará el valor que le corresponde a cada clave
#y será más fácil obtener lo que necesito. 
#En este ejemplo, Python asocia la clave A con 0 y 3,
#por lo que su media es 1.5
#La clave B es asociada con 1 y 4, por lo que
#su media es 2.5
#La clave C es asociada con 2 y 5, por lo que
#su media es 3.5
```

# Manipulación de Data Frames

```python
df3 = pd.DataFrame({
    'key': ['A', 'B', 'C','A', 'B', 'C'],
    'value1' : range(6),
    'value2' : np.random.randint(0,100,6) #números aleatorios entre 0 y 100. Exactamente 6 números.
})
df3
```

## Aggregate

```python
df3.groupby('key').aggregate(['min', np.median, np.mean, max])
#Las funciones de Python se escriben entre comillas o al colocarlas se tornarán 
#de color verde. 
#Para las de una librería, se ponen con su acrónimo seguido del punto
```

```python
df3.groupby('key').aggregate({
    'value1' : 'min',
    'value2' : 'max'
})
```

## Filter

```python
df3.groupby('key').std() #Desviación estándar
```

```python
def filter_func(x):
    return x['value2'].std() > 10
```

```python
display("df3", "df3.groupby('key').std()", "df3.groupby('key').filter(filter_func)")
```

```python
df3.groupby('key').filter(filter_func)
```

## Transform

#### Esta función se usa cuando quiero obtener una versión reducida de los datos pero sin combinar.
#### Se conserva la estructura original.

```python
df3.groupby('key').transform(lambda x : x - x.mean())
```

# Apply

### Nos deja aplicar una función arbitraria a un DF

```python
def norm_by_col2(x):
    x['value1'] /= x['value2'].sum() 
    return x
# Hacer /= es exactamente lo mismo que poner
#x['values1'] = x['values1'] / x['values2'].sum()
#Es, por así decirlo, la versión reducida.
```

```python
df3.groupby('key').apply(norm_by_col2)
```

# División de datos por columnas

```python
L = [0,1,0,1,2,0]
print(df3)
df3.groupby(L).sum()
```

### Con esto, puedo asignar nuevas "claves" a los valores de mi data frame y realizar operaciones con ellos, en este ejemplo, a los elementos que se encuentran en la posición 0, 2 y 5 les asigno la clave 0. A los elementos de las posiciones 1 y 3 les asigno el valor 1 y, finalmente, al elemento de la posición 4 le asigno la clave 1.
### Así, aplico la función sum con esta nueva agrupación.

```python
df3.groupby(df3['key']).sum()
```

# Gráficos desde Data Frames y exportación

```python
years = [year for year in range (1900, 2020)]
deads = [(y + np.random.uniform(0,100) - 1850) for y in years]
```

```python
dfr = pd.DataFrame({
    'year' : years,
    'deads' : deads
})
dfr.head()
```

```python
dfr.plot(kind = "scatter", x = "year", y = "deads", figsize = (10, 8))
```

```python
my_plot = dfr.plot(x = "year", y = "deads", figsize = (10, 8))
```

```python
my_fig = my_plot.get_figure()
```

```python
my_fig.savefig("data_frame_plot.png")
```
