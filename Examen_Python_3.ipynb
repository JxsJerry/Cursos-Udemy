```python
n = 256

X = np.linspace(-np.pi, np.pi, n, endpoint=True)

Y = np.sin(2 * X) 

plt.plot(X, Y + 1, color='blue', alpha=1.00)

plt.plot(X, Y - 1, color='blue', alpha=1.00)

plt.fill_between(X, 1, Y+1, alpha = 0.3) #Con esto lleno el espacio
#que existe entre los valores de X tomados como si de una línea recta se tratara
#y los valores generados por Y+1 de color azul.

plt.fill_between(X, -1, Y-1, where = -1 < Y-1, color='blue',alpha = 0.3)
#Con esto lleno el espacio que existe entre los valores de X 
#tomados como si de una línea recta se tratara y los valores generados 
#por -1 y Y-1 de color azul. Con la condición de que los valores
#que obtengo de hacer Y-1 sean mayores que -1.

plt.fill_between(X, -1, Y-1, where = Y-1 < -1, color = "red", alpha = 0.3)
#Con esto lleno el espacio que existe entre los valores de X 
#tomados como si de una línea recta se tratara y los valores generados 
#por -1 y Y-1 de color rojo. Con la condición de que los valores
#que obtengo de hacer Y-1 sean menores que -1.
```

```python
n = 1024

X = np.random.normal(0,1,n)

Y = np.random.normal(0,1,n) 

theta = np.arctan2(Y,X) #Con la función arctan2 se escoge
#Adecuadamente el valor del arco tangente en relación con el 
#cuadrante en el que se evalua.

colors = theta

plt.axes([0.025, 0.025, 0.95, 0.95])

plt.scatter(X,Y, s=75, c=colors,alpha=0.75)

plt.xlim(-1.5, 1.5)

plt.ylim(-1.5, 1.5)
```

```python
n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n) 
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white') 
for x, y in zip(X, Y1):     
    plt.text(x + 0.05, y + 0.01, '%.2f' % y, ha='center', va='bottom')
for x, y in zip(X, -Y2):
    plt.text(x + 0.05, y - 0.15 , '%.2f' % np.absolute(y), ha='center', va='bottom')

plt.ylim(-1.25, +1.25)
```

```python
n = 256

X = np.linspace(-np.pi, np.pi, n, endpoint=True)

Y = np.sin(2 * X)



plt.axes([0.025, 0.025, 0.95, 0.95])



plt.plot(X, Y + 1, color='blue', alpha=1.00)

plt.fill_between(X, 1, Y + 1, color='blue', alpha=.25)



plt.plot(X, Y - 1, color='blue', alpha=1.00)

plt.fill_between(X, -1, Y - 1, (Y - 1) > -1, color='blue', alpha=.25)

plt.fill_between(X, -1, Y - 1, (Y - 1) < -1, color='red',  alpha=.25)



plt.xlim(-np.pi, np.pi)

plt.xticks(())

plt.ylim(-2.5, 2.5)

plt.yticks(())



plt.show()
```

```python
n = 1024

X = np.random.normal(0, 1, n)

Y = np.random.normal(0, 1, n)

T = np.arctan2(Y, X) 



plt.axes([0.025, 0.025, 0.95, 0.95])

plt.scatter(X, Y, s=75, c=T, alpha=.5)



plt.xlim(-1.5, 1.5)

plt.xticks(())

plt.ylim(-1.5, 1.5)

plt.yticks(())



plt.show()
```

```python
n = 12

X = np.arange(n)

Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)



plt.axes([0.025, 0.025, 0.95, 0.95])

plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')

plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')



for x, y in zip(X, Y1):

    plt.text(x + 0.4, y + 0.05, '%.2f' % y, ha='center', va= 'bottom')



for x, y in zip(X, Y2):

    plt.text(x + 0.4, -y - 0.05, '%.2f' % y, ha='center', va= 'top')



plt.xlim(-.5, n)

plt.xticks(())

plt.ylim(-1.25, 1.25)

plt.yticks(())



plt.show()
```

```python
n = 20

Z = np.ones(n)

Z[-1] *= 2



plt.axes([0.025, 0.025, 0.95, 0.95])



plt.pie(Z, explode=Z*.05, colors = ['%f' % (i/float(n)) for i in range(n)])

plt.axis('equal')

plt.xticks(())

plt.yticks()



plt.show()
```
