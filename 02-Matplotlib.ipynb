# Matplotlib

### Con esto puedo hacer que los graficos que genere se muestren en pantalla

```python
%matplotlib inline 
```

```python
import matplotlib.pyplot as plt # Con esto, importo la librería pyplot y le doy el alias plt
```

```python
x = (1,2,3,4)
```

```python
plt.plot(x)
```

#### En Python, los puntos son unidos de entrada a diferencia de R que solo genera los puntos.

```python
plt.plot(x)
plt.xlabel("Eje de abcisas")
plt.ylabel("Eje de ordenadas")
plt.show() #Esto para mostrar el gráfico en cuestión.
```

#### Cuando solo introduzco un valor, los elementos que se encuentran dentro del gráfico Python los entiende, al igual que R, como los elementos que corresponden al eje de las Y. Además, el eje X tendrá la misma cantidad de elementos que el del valor suministrado, es decir, si ingreso 4 elementos para el eje de las Y, Python añadirá al eje X cuatro elementos comenzando a partir del 0 pues es como Python cuenta.

```python
x = (1,2,3,4)
y = (1,4,9,16)
plt.plot(x,y)
```

#### IMPORTANTE: La cardinalidad de los vectores que se suministran deben ser iguales.

```python
plt.plot(x,y,"ro")
```

#### Como Python heredó los códigos plot de MatLab, para generar diversos tipos de gráficos, debemos añadir parámetros adicionales en formato String que representen el color y la figura con la que se quieren representar. (En ese caso "r" para red y "o" para círculos)

```python
plt.plot(x,y,"ro")
plt.axis([0,6, 0, 20]) #Con esto modifico los límites de los ejes.
plt.show()
```

#### Existen otros comandos que también modifican los puntos del gráfico, como por ejemplo: +, ,, ., 1, 2, 3, 4

```python
plt.plot(x,y,linewidth=2.0) #Con este comando, modifico el ancho de la línea del gráfico.
```

```python
line, =plt.plot(x,y,"-")
line.set_antialiased(False)
```

#### Además de la "-" también podemos colocar otras maneras de unir los puntos del gráfico, como por ejemplo: "--", "-.", "..", "steps"

```python
lines = plt.plot(data, data, data, data**2)
plt.setp(lines, color = "r", linewidth = 2.0)
```

#### Configura una lista de propiedades múltiples para un conjunto de líneas. 

```python
lines = plt.plot(data, data, data, data**2)
plt.setp(lines, "color", "r", "linewidth", 2.0)
```

```python
def f(x):
    return np.exp(-x)*np.cos(2*np.pi*x)
```

```python
x1 = np.arange(0, 5, 0.1)
x2 = np.arange(0, 5, 0.2)
```

```python
plt.figure(1) 
plt.subplot(211) #El número 211 indica que en una matriz de 2 filas y una columna, el primer 
# elemento será el gráfico 1.
plt.plot(x1, f(x1), "ro", x2, f(x2), "k") #Con esta notación, x1 y x2 fungen como los valores del eje X y 
# f(x1) y f(x2) como los valores del eje Y.

plt.subplot(212) #El número 212 indica que el segundo elemento de la matriz será el segundo gráfico generado.
plt.plot(x2, f(x2), "g--")

plt.show()
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot([1,2,3])
plt.subplot(2,1,2)
plt.plot([4,5,6])

plt.figure(2)
plt.plot([4,5,6])

plt.figure(1) #Esto es para cambiar entre figuras, en este caso, se mueve a la figura 1, que
# corresponde a la matriz que contiene a las dos primeras gráficas.
plt.subplot(211)
plt.title("Esto es el primer título")
```

# Textos en los gráficos

```python
mu = 100
sigma = 20

x = mu + sigma * np.random.randn(10000)
```

```python
n, bins, patches = plt.hist(x, 50, normed = 1, facecolor = "g", alpha = 0.6)

plt.xlabel("Cociente Intelectual", fontsize = 14, color = "green")
plt.ylabel("Probabilidad")
plt.title(r"Histograma de CI $N(\mu, \sigma)$")
plt.text(40, 0.015, r'$\mu = 100, \ \sigma = 20$')
#Con la letra r designamos los siguientes valores como tipo "row"
plt.axis([20,180,0,0.025])
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize = (10,6), dpi = 90)
plt.subplot(111)

x = np.arange(-2*np.pi, 2*np.pi, 0.01)
y = np.cos(x)
plt.plot(x,y, lw = 2.0)

plt.annotate("Máximo local", xy=(4*np.pi, 1), xytext = (15, 1.5),
            arrowprops = dict(facecolor = "black", shrink = 0.05))
plt.ylim(-2,2)
plt.show()
```

```python
from matplotlib.ticker import NullFormatter
```

```python
mu = 0.5
sd = 0.3
y = mu + sd*np.random.randn(1000)
y = y[(y>0) & (y<1)]
y.sort() #Los ordenamos de forma creciente
x = np.arange(len(y))
```

```python
plt.figure(figsize=(10, 8))

plt.subplot(221)
plt.plot(x,y)
plt.yscale("linear")
plt.xscale("linear")
plt.title("Escala lineal")
plt.grid(True)


plt.subplot(222)
plt.plot(x,y)
plt.yscale("log")
plt.title("Escala semilogarítmica")
plt.grid(True)

plt.subplot(223)
plt.plot(x,y - y.mean()) #Este gráfico (symetriclog) tiene sentido si 
#graficamos x e y - y.mean()
plt.yscale("symlog", linthreshy=0.01) #Las divisiones semilogarítmicas
#las quiero cada 0.01
plt.title("Escala log simétrico")
plt.grid(True)

plt.subplot(224)
plt.plot(x,y)
plt.yscale("logit")
plt.title("Escala logística")
plt.gca().yaxis.set_minor_formatter(NullFormatter())
plt.grid(True)

plt.subplots_adjust(top = 0.92, bottom = 0.08, left = 0.1, right = 0.95, hspace = 0.35, wspace = 0.35)
plt.show()
```

# Cambios en los ejes de dibujo

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint = True) #Con el 
#parámetro endpoint=True hacemos que el último valor sí se tome en cuenta.
S, C = np.sin(x), np.cos(x)
```

```python
plt.plot(x, S, color = "blue", linewidth = 1.2, linestyle = "-")
plt.plot(x, C, color = "green", linewidth = 1.2, linestyle = "-")
plt.xlim(-4,4)
plt.ylim(-1.2,1.2)
plt.xticks(np.linspace(-4, 4, 9, endpoint=True))
plt.yticks(np.linspace(-1, 1, 9, endpoint=True))
plt.show()
```

```python
plt.plot(x, S, color = "blue", linewidth = 1.2, linestyle = "-")
plt.plot(x, C, color = "green", linewidth = 1.2, linestyle = "-")
plt.xlim(x.min()*1.1, x.max()*1.1) #Con este comando los límites
#de mi gráfico se estiran o contraen de manera dinámica tomando
#en cuenta el valor mínimo y máximo de las x's tomadas en cuenta y agrandándolo
#o contrayéndolo en cierto porcentaje
plt.ylim(S.min()*1.1, S.max()*1.1) #Con este comando los límites
#de mi gráfico se estiran o contraen de manera dinámica tomando
#en cuenta el valor mínimo y máximo de cierta función y agrandándolo
#o contrayéndolo en cierto porcentaje
plt.xticks(np.linspace(-4, 4, 9, endpoint=True))
plt.yticks(np.linspace(-1, 1, 9, endpoint=True))
plt.show()
```

```python
plt.plot(x, S, color = "blue", linewidth = 1.2, linestyle = "-")
plt.plot(x, C, color = "green", linewidth = 1.2, linestyle = "-")
plt.xlim(x.min()*1.1, x.max()*1.1) 
plt.ylim(S.min()*1.1, S.max()*1.1) 
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], 
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$']) #También 
#se pueden generar los valores mostrados generados en los ejes 
#nombrándolos manualmente con un vector.
#Además, para poner los nombres en formato LaTeX los puedo asignar 
#añadiendo un vector adicional.
plt.yticks(np.linspace(-1, 1, 3, endpoint=True))

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none') #Con estos comandos quitamos los
#bordes derecho y superior del dibujo (gráfica)

plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.plot(x, S, color = "blue", linewidth = 1.2, linestyle = "-", label = "Seno")
plt.plot(x, C, color = "green", linewidth = 1.2, linestyle = "-", label = "Coseno")
plt.xlim(x.min()*1.1, x.max()*1.1) 
plt.ylim(S.min()*1.1, S.max()*1.1) 
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], 
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$']) #También 
#se pueden generar los valores mostrados generados en los ejes 
#nombrándolos manualmente con un vector.
#Además, para poner los nombres en formato LaTeX los puedo asignar 
#añadiendo un vector adicional.
plt.yticks(np.linspace(-1, 1, 2, endpoint=True), 
          [-1, 1])

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none') 
ax.xaxis.set_ticks_position('bottom') #Con esto hacemos que
#el eje de las x quede en la parte inferior. Esto para la siguiente instrucción.
ax.spines['bottom'].set_position(('data', 0)) #Con esto, movemos el
#eje de las x a la posción 0 (al centro del gráfico).
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0)) #Con esto, movemos el
#eje de las y a la posción 0 (al centro del gráfico).

plt.legend(loc="upper left")

x0 = 2*np.pi/3
plt.plot([x0,x0], [0, np.sin(x0)], color = "blue", linewidth = 2.5, linestyle = "--") 
#Una línea que va de (x0,0) hasta (x0,sen(x0))
plt.scatter([x0, ], [np.sin(x0), ], 50, color = "blue")
#Trazo un punto de color azul
plt.annotate(r'$\sin(\frac{2\pi}{3}) = \frac{\sqrt{3}}{2}$',
            xy = (x0, np.sin(x0)), xycoords = "data",
            xytext = (+20,+40), textcoords = "offset points", 
            fontsize = 16, arrowprops = dict(arrowstyle = "->", connectionstyle="arc3, rad=0.2"))
#Añadimos texto con formato LaTeX y una flecha que señala el punto.
plt.plot([x0,x0], [0, np.cos(x0)], color = "green", linewidth = 2.5, linestyle = "--") 
#Una línea que va de (x0,0) hasta (x0,sen(x0))
plt.scatter([x0, ], [np.cos(x0), ], 50, color = "green")
#Trazo un punto de color azul
plt.annotate(r'$\cos(\frac{2\pi}{3}) = -\frac{1}{2}$',
            xy = (x0, np.cos(x0)), xycoords = "data",
            xytext = (-90,-60), textcoords = "offset points", 
            fontsize = 16, arrowprops = dict(arrowstyle = "->", connectionstyle="arc3, rad=0.2"))
#Añadimos texto con formato LaTeX y una flecha que señala el punto.

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor="white", edgecolor="None", alpha=0.6))

plt.show()
```
