# Matplotlib

### Con esto puedo hacer que los graficos que genere se muestren en pantalla

```python
%matplotlib inline 
```

```python
import matplotlib.pyplot as plt # Con esto, importo la librer√≠a pyplot y le doy el alias plt
```

```python
x = (1,2,3,4)
```

```python
plt.plot(x)
```

#### En Python, los puntos son unidos de entrada a diferencia de R que solo genera los puntos.

```python
plt.plot(x)
plt.xlabel("Eje de abcisas")
plt.ylabel("Eje de ordenadas")
plt.show() #Esto para mostrar el gr√°fico en cuesti√≥n.
```

#### Cuando solo introduzco un valor, los elementos que se encuentran dentro del gr√°fico Python los entiende, al igual que R, como los elementos que corresponden al eje de las Y. Adem√°s, el eje X tendr√° la misma cantidad de elementos que el del valor suministrado, es decir, si ingreso 4 elementos para el eje de las Y, Python a√±adir√° al eje X cuatro elementos comenzando a partir del 0 pues es como Python cuenta.

```python
x = (1,2,3,4)
y = (1,4,9,16)
plt.plot(x,y)
```

#### IMPORTANTE: La cardinalidad de los vectores que se suministran deben ser iguales.

```python
plt.plot(x,y,"ro")
```

#### Como Python hered√≥ los c√≥digos plot de MatLab, para generar diversos tipos de gr√°ficos, debemos a√±adir par√°metros adicionales en formato String que representen el color y la figura con la que se quieren representar. (En ese caso "r" para red y "o" para c√≠rculos)

```python
plt.plot(x,y,"ro")
plt.axis([0,6, 0, 20]) #Con esto modifico los l√≠mites de los ejes.
plt.show()
```

```python
import numpy as np
```

```python
data = np.arange(0.0, 10.0, 0.2)
data
```

```python
plt.plot(data, data, "r--", data, data**2, "bs", data, data**3, "g^")
plt.show()
```

#### Existen otros comandos que tambi√©n modifican los puntos del gr√°fico, como por ejemplo: +, ,, ., 1, 2, 3, 4

```python
plt.plot(x,y,linewidth=2.0) #Con este comando, modifico el ancho de la l√≠nea del gr√°fico.
```

```python
line, =plt.plot(x,y,"-")
line.set_antialiased(False)
```

#### Adem√°s de la "-" tambi√©n podemos colocar otras maneras de unir los puntos del gr√°fico, como por ejemplo: "--", "-.", "..", "steps"

```python
lines = plt.plot(data, data, data, data**2)
plt.setp(lines, color = "r", linewidth = 2.0)
```

#### Configura una lista de propiedades m√∫ltiples para un conjunto de l√≠neas. 

```python
lines = plt.plot(data, data, data, data**2)
plt.setp(lines, "color", "r", "linewidth", 2.0)
```

#### Otra manera de modificar los elementos del gr√°fico.

```python
plt.plot(x,y,alpha = 0.2) #Este comando baja la opacidad del gr√°fico
```

```python
plt.plot(x,y,marker="+", linestyle = ":")
```

### Para acceder a la documentaci√≥n, basta con ingresar el c√≥digo plt.setp("nombre del gr√°fico anteriormente generado") y entonces Python arrojar√° los par√°metros que puede incluir el c√≥digo.

```python
plt.setp(lines)
```

```python
def f(x):
    return np.exp(-x)*np.cos(2*np.pi*x)
```

```python
x1 = np.arange(0, 5, 0.1)
x2 = np.arange(0, 5, 0.2)
```

```python
plt.figure(1) 
plt.subplot(211) #El n√∫mero 211 indica que en una matriz de 2 filas y una columna, el primer 
# elemento ser√° el gr√°fico 1.
plt.plot(x1, f(x1), "ro", x2, f(x2), "k") #Con esta notaci√≥n, x1 y x2 fungen como los valores del eje X y 
# f(x1) y f(x2) como los valores del eje Y.

plt.subplot(212) #El n√∫mero 212 indica que el segundo elemento de la matriz ser√° el segundo gr√°fico generado.
plt.plot(x2, f(x2), "g--")

plt.show()
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot([1,2,3])
plt.subplot(2,1,2)
plt.plot([4,5,6])

plt.figure(2)
plt.plot([4,5,6])

plt.figure(1) #Esto es para cambiar entre figuras, en este caso, se mueve a la figura 1, que
# corresponde a la matriz que contiene a las dos primeras gr√°ficas.
plt.subplot(211)
plt.title("Esto es el primer t√≠tulo")
```

# Textos en los gr√°ficos

```python
mu = 100
sigma = 20

x = mu + sigma * np.random.randn(10000)
```

```python
n, bins, patches = plt.hist(x, 50, normed = 1, facecolor = "g", alpha = 0.6)

plt.xlabel("Cociente Intelectual", fontsize = 14, color = "green")
plt.ylabel("Probabilidad")
plt.title(r"Histograma de CI $N(\mu, \sigma)$")
plt.text(40, 0.015, r'$\mu = 100, \ \sigma = 20$')
#Con la letra r designamos los siguientes valores como tipo "row"
plt.axis([20,180,0,0.025])
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize = (10,6), dpi = 90)
plt.subplot(111)

x = np.arange(-2*np.pi, 2*np.pi, 0.01)
y = np.cos(x)
plt.plot(x,y, lw = 2.0)

plt.annotate("M√°ximo local", xy=(4*np.pi, 1), xytext = (15, 1.5),
            arrowprops = dict(facecolor = "black", shrink = 0.05))
plt.ylim(-2,2)
plt.show()
```

```python
from matplotlib.ticker import NullFormatter
```

```python
mu = 0.5
sd = 0.3
y = mu + sd*np.random.randn(1000)
y = y[(y>0) & (y<1)]
y.sort() #Los ordenamos de forma creciente
x = np.arange(len(y))
```

```python
plt.figure(figsize=(10, 8))

plt.subplot(221)
plt.plot(x,y)
plt.yscale("linear")
plt.xscale("linear")
plt.title("Escala lineal")
plt.grid(True)


plt.subplot(222)
plt.plot(x,y)
plt.yscale("log")
plt.title("Escala semilogar√≠tmica")
plt.grid(True)

plt.subplot(223)
plt.plot(x,y - y.mean()) #Este gr√°fico (symetriclog) tiene sentido si 
#graficamos x e y - y.mean()
plt.yscale("symlog", linthreshy=0.01) #Las divisiones semilogar√≠tmicas
#las quiero cada 0.01
plt.title("Escala log sim√©trico")
plt.grid(True)

plt.subplot(224)
plt.plot(x,y)
plt.yscale("logit")
plt.title("Escala log√≠stica")
plt.gca().yaxis.set_minor_formatter(NullFormatter())
plt.grid(True)

plt.subplots_adjust(top = 0.92, bottom = 0.08, left = 0.1, right = 0.95, hspace = 0.35, wspace = 0.35)
plt.show()
```

# Cambios en los ejes de dibujo

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint = True) #Con el 
#par√°metro endpoint=True hacemos que el √∫ltimo valor s√≠ se tome en cuenta.
S, C = np.sin(x), np.cos(x)
```

```python
plt.plot(x, S, color = "blue", linewidth = 1.2, linestyle = "-")
plt.plot(x, C, color = "green", linewidth = 1.2, linestyle = "-")
plt.xlim(-4,4)
plt.ylim(-1.2,1.2)
plt.xticks(np.linspace(-4, 4, 9, endpoint=True))
plt.yticks(np.linspace(-1, 1, 9, endpoint=True))
plt.show()
```

```python
plt.plot(x, S, color = "blue", linewidth = 1.2, linestyle = "-")
plt.plot(x, C, color = "green", linewidth = 1.2, linestyle = "-")
plt.xlim(x.min()*1.1, x.max()*1.1) #Con este comando los l√≠mites
#de mi gr√°fico se estiran o contraen de manera din√°mica tomando
#en cuenta el valor m√≠nimo y m√°ximo de las x's tomadas en cuenta y agrand√°ndolo
#o contray√©ndolo en cierto porcentaje
plt.ylim(S.min()*1.1, S.max()*1.1) #Con este comando los l√≠mites
#de mi gr√°fico se estiran o contraen de manera din√°mica tomando
#en cuenta el valor m√≠nimo y m√°ximo de cierta funci√≥n y agrand√°ndolo
#o contray√©ndolo en cierto porcentaje
plt.xticks(np.linspace(-4, 4, 9, endpoint=True))
plt.yticks(np.linspace(-1, 1, 9, endpoint=True))
plt.show()
```

```python
plt.plot(x, S, color = "blue", linewidth = 1.2, linestyle = "-")
plt.plot(x, C, color = "green", linewidth = 1.2, linestyle = "-")
plt.xlim(x.min()*1.1, x.max()*1.1) 
plt.ylim(S.min()*1.1, S.max()*1.1) 
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], 
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$']) #Tambi√©n 
#se pueden generar los valores mostrados generados en los ejes 
#nombr√°ndolos manualmente con un vector.
#Adem√°s, para poner los nombres en formato LaTeX los puedo asignar 
#a√±adiendo un vector adicional.
plt.yticks(np.linspace(-1, 1, 3, endpoint=True))

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none') #Con estos comandos quitamos los
#bordes derecho y superior del dibujo (gr√°fica)

plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.plot(x, S, color = "blue", linewidth = 1.2, linestyle = "-", label = "Seno")
plt.plot(x, C, color = "green", linewidth = 1.2, linestyle = "-", label = "Coseno")
plt.xlim(x.min()*1.1, x.max()*1.1) 
plt.ylim(S.min()*1.1, S.max()*1.1) 
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], 
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$']) #Tambi√©n 
#se pueden generar los valores mostrados generados en los ejes 
#nombr√°ndolos manualmente con un vector.
#Adem√°s, para poner los nombres en formato LaTeX los puedo asignar 
#a√±adiendo un vector adicional.
plt.yticks(np.linspace(-1, 1, 2, endpoint=True), 
          [-1, 1])

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none') 
ax.xaxis.set_ticks_position('bottom') #Con esto hacemos que
#el eje de las x quede en la parte inferior. Esto para la siguiente instrucci√≥n.
ax.spines['bottom'].set_position(('data', 0)) #Con esto, movemos el
#eje de las x a la posci√≥n 0 (al centro del gr√°fico).
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0)) #Con esto, movemos el
#eje de las y a la posci√≥n 0 (al centro del gr√°fico).

plt.legend(loc="upper left")

x0 = 2*np.pi/3
plt.plot([x0,x0], [0, np.sin(x0)], color = "blue", linewidth = 2.5, linestyle = "--") 
#Una l√≠nea que va de (x0,0) hasta (x0,sen(x0))
plt.scatter([x0, ], [np.sin(x0), ], 50, color = "blue")
#Trazo un punto de color azul
plt.annotate(r'$\sin(\frac{2\pi}{3}) = \frac{\sqrt{3}}{2}$',
            xy = (x0, np.sin(x0)), xycoords = "data",
            xytext = (+20,+40), textcoords = "offset points", 
            fontsize = 16, arrowprops = dict(arrowstyle = "->", connectionstyle="arc3, rad=0.2"))
#A√±adimos texto con formato LaTeX y una flecha que se√±ala el punto.
plt.plot([x0,x0], [0, np.cos(x0)], color = "green", linewidth = 2.5, linestyle = "--") 
#Una l√≠nea que va de (x0,0) hasta (x0,sen(x0))
plt.scatter([x0, ], [np.cos(x0), ], 50, color = "green")
#Trazo un punto de color azul
plt.annotate(r'$\cos(\frac{2\pi}{3}) = -\frac{1}{2}$',
            xy = (x0, np.cos(x0)), xycoords = "data",
            xytext = (-90,-60), textcoords = "offset points", 
            fontsize = 16, arrowprops = dict(arrowstyle = "->", connectionstyle="arc3, rad=0.2"))
#A√±adimos texto con formato LaTeX y una flecha que se√±ala el punto.

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor="white", edgecolor="None", alpha=0.6))

plt.show()
```
