# Cálculo de Primas de Seguros

```python
import pandas as pd
import numpy as np
```

```python
tabmort = pd.read_csv("Tabla_Actuarial.csv", encoding = 'unicode_escape')
tabmort.head()
```

```python
A = int(tabmort.loc[tabmort.x == 20,'Mx'])
B = int(tabmort.loc[tabmort.x == 20 + 10,'Mx'])
A/B
#Nota: convertir a número para evitar NaN's.
```

```python
tipo = str(input("Introduce el tipo de seguro que quieras calcular: ")) #El usuario introduce el tipo de seguro.
tipom = tipo.lower() #Por cualquier cosa, convierto el texto a minúsculas para trabajarlo mejor.
tipoS = tipom.capitalize() #Se transforma la primera letra del texto ingresado a mayúscula.
if tipoS == "Vitalicio": #La entrada es igual a Vitalicio.
    y = int(input("Introduce la edad del asegurado: ")) #Edad transformada a entero.
    SA = int(input("Introduce la Suma Asegurada deseada: ")) #Suma asegurada transformada a entero.
    if y >= 16: #Prueba de edad.
        A = ((float(tabmort.loc[tabmort.x == y,'Mx'])/float(tabmort.loc[tabmort.x == y,'Dx'])))*SA #Transformado a tipo float.
    else: #En caso de no cumplirse que la edad entre dentro del rango de la tabla.
        A = "No es posible calcular el seguro" 
elif tipoS == "Temporal": #La entrada es igual a Temporal.
    y = int(input("Introduce la edad del asegurado: ")) #Edad transformada a entero.
    t =  int(input("Introduce la temporalidad de cobertura: ")) #Temporalidad transformada a entero.
    SA = int(input("Introduce la Suma Asegurada deseada: ")) #Suma asegurada transformada a entero.
    if y >= 16 and y + t < 99: #Prueba para la edad y la temporalidad.
        A = ((float(tabmort.loc[tabmort.x == y,'Mx'])- float(tabmort.loc[tabmort.x == y + t,'Mx']))
             /float(tabmort.loc[tabmort.x == y,'Dx']))*SA #Transformado a tipo float.
    else: #En caso de no cumplirse que la edad y su suma con la temporalidad entre dentro del rango de la tabla.
        A = "No es posible calcular el seguro" 
elif tipoS == "Dotal": #La entrada es igual a Dotal.
    y = int(input("Introduce la edad del asegurado: ")) #Edad transformada a entero.
    t =  int(input("Introduce la temporalidad de cobertura: ")) #Temporalidad transformada a entero.
    SA = int(input("Introduce la Suma Asegurada deseada: ")) #Suma asegurada transformada entero.
    if y >= 16 and y + t < 99: #Prueba para la edad y la temporalidad.
        A = ((float(tabmort.loc[tabmort.x == y,'Mx'])- float(tabmort.loc[tabmort.x == y + t,'Mx'])
              + float(tabmort.loc[tabmort.x == y + t,'Dx']))
             /float(tabmort.loc[tabmort.x == y,'Dx']))*SA #Transformado a tipo float.
    else: #En caso de no cumplirse que la edad y su suma con la temporalidad entre dentro del rango de la tabla.
        A = "No es posible calcular el seguro" 
else:
    A = "No existe la modalidad introducida"
print("La prima de tu seguro es: %s" %A)
```
